# input from vgg16

```python
vgg16_model=VGG16(weights='imagenet',include_top=False,input_shape=(224,224,3))
vgg16_model.summary()
```

```python
from keras.layers import Dense,Flatten
```

```python
for layer in vgg16_model.layers:
  layer.trainable=False  # all parameters and layers are fixed cant train it
```

## output of Flatten and dense

```python
x=Flatten()(vgg16_model.output)
prediction=Dense(3,activation='softmax')(x)  # 3 classes rock shah tom
```

```python
from tensorflow.keras.models import Model
```

```python
model=Model(inputs=vgg16_model.inputs,outputs=prediction)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```
