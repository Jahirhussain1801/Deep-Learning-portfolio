```python
def create_model():
  model=Sequential()

  model.add(Dense(units=20,input_dim=8,activation='relu',kernel_initializer="glorot_uniform"))
  model.add(Dense(units=60,activation='relu',kernel_initializer="glorot_uniform"))
  model.add(Dense(units=50,activation='relu',kernel_initializer="glorot_uniform"))
  model.add(Dense(units=1,activation='sigmoid',kernel_initializer='glorot_uniform'))

  model.summary()

  model.compile(optimizer='adam',loss="binary_crossentropy",metrics=["binary_accuracy"])

  return model
```

```python
clf=KerasClassifier(build_fn=create_model)
```

```python
grid=GridSearchCV(estimator=clf,param_grid={"batch_size":[10,20,40],"epochs":[10,15,20]})
```

```python
def create_model(neuron1,neuron2,activation_function,init,learning_rate,dropout_rate):

  model=Sequential()

  model.add(Dense(units=neuron1,input_dim=8,activation=activation_function,kernel_initializer=init))
  model.add(Dropout(dropout_rate))

  model.add(Dense(units=neuron2,activation=activation_function,kernel_initializer=init))
  model.add(Dropout(dropout_rate))

  model.add(Dense(units=1,activation='sigmoid',kernel_initializer='glorot_uniform'))

  model.summary()

  adam=tf.keras.optimizers.Adam(learning_rate=learning_rate,beta_1=0.5,beta_2=0.96,epsilon=1e-07)

  model.compile(optimizer=adam,loss="binary_crossentropy",metrics=["binary_accuracy"])

  return model
```
