```python
x_train=x_train/255
x_test=x_test/255   # normalization to btw 0 to 1 as bl/wh is 0 to 255 so / by 255 to get value b/w 0 to 1
```

```python
x_train[0]
```

```python
# 1 build the architecture
# 2 model compilation
# 3 model training
# 4 model testing
```

```python
from keras.models import Sequential
from keras.layers import Dense,Flatten
```

```python
model=Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(units=50,activation='relu'))  # 50 bias added extra
model.add(Dense(units=10,activation='relu'))
model.add(Dense(units=10,activation='softmax'))

model.summary()
```

```python
model.compile(optimizer='adam', loss="sparse_categorical_crossentropy", metrics=["sparse_categorical_accuracy"])
```

**training the model**

```python
model_training=model.fit(x=x_train,y=y_train,batch_size=10,epochs=20,verbose='auto')
```

# predicting the model

```python
y_pred=model.predict(x_test)
```
