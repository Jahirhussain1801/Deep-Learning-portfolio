```python
from tensorflow import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Conv2D,AveragePooling2D,Flatten,Dense
import numpy as np
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data(path="minst.npz")
```

```python
x_train.shape
```

```python
x_train=np.pad(array=x_train,pad_width=((0,0),(1,1),(1,1))) # or (0,0),(2,2),(2,2)
```

```python
x_train.shape
```

```python
x_test=np.pad(array=x_test,pad_width=((0,0),(2,2),(2,2)))
```

```python
x_test.shape
```

```python
import pandas as pd
pd.DataFrame(x_train[3000])
```

```python
y_train[3000]
```

```python
x_train1=x_train/255
x_test1=x_test/255
```

```python
pd.DataFrame(x_train1[3000])
```

```python
model=Sequential()

model.add(Conv2D(input_shape=(32,32,1),filters=6,kernel_size=(5,5),strides=1,activation='tanh'))
model.add(AveragePooling2D(pool_size=(2,2),strides=2))

model.add(Conv2D(filters=16,kernel_size=(5,5),strides=1,activation='tanh'))
model.add(AveragePooling2D(pool_size=(2,2),strides=2))

model.add(Conv2D(filters=120,kernel_size=(5,5),strides=1,activation='tanh'))

model.add(Flatten())

model.add(Dense(units=84,activation='tanh'))

model.add(Dense(units=10,activation='softmax'))



model.summary()
```

```python
model.compile(optimizer='rmsprop',loss='sparse_categorical_crossentropy',metrics=['sparse_categorical_accuracy'])
```

```python
model_train=model.fit(x=x_train1,y=y_train,epochs=15,validation_data=(x_test1,y_test))
```

```python
model.evaluate(x_test1,y_test)
```

```python
predict=model.predict(x_test1)
```
